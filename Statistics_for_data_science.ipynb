```python
df = pd.read_csv("Laliga.csv")
df.head(8)
```

```python
def replaceDash(col):
  df[col] = df[col].str.replace("-", "0")
cols = ['Points','GamesPlayed','GamesWon','GamesDrawn','GamesLost','GoalsFor','GoalsAgainst','Champion','Runner-up','Third','Fourth','Fifth','Sixth','T']
for col in cols:
  replaceDash(col)
df
```

**Print all the teams which have started playing between 1930-1980.**

**Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference**

```python
def Goal_diff_count():   
     df['Goal_diff_count'] = df['GoalsFor'].astype(int)-df['GoalsAgainst'].astype(int)
     # if require to get the absolute value below line can be used
     # df['Goal_diff_count'] = df['Goal_diff_count'].abs()
     return df[['Team','Goal_diff_count']].sort_values(by = 'Goal_diff_count',ascending=False)
    
Goal_diff_count()
```

```python
print("The team which has maximum goal difference is 👇")
Goal_diff_count().head(1)
```

```python
print("The team which has minimum goal difference is 👇")
Goal_diff_count().head(1)
```

**Create a new column with name “Winning Percent” and append it to the data set**

```python
    
df['Winning Percent'] = (df['GamesWon'].astype(int) / df['GamesPlayed'].astype(int) )*100
df['Winning Percent'].fillna(0,inplace = True)
df[['Team','Winning Percent']].head()
```

**Group teams based on their “Best position” and print the sum of their points for all positions**

```python
df[["Points", "BestPosition"]] = df[["Points", "BestPosition"]].apply(pd.to_numeric)
df.dtypes
```

```python
groupOfBestPos = df[['Team','Points','BestPosition']].groupby('BestPosition')
groupOfBestPos.sum()
```
